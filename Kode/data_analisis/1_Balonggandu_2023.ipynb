```python
import os
import numpy as np
import pandas as pd
```

```python
def read_excel(path):
    return pd.read_excel(path)
```

```python
path_1 = '../../Data/Preprocessing/Data Balonggandu/2023/2_input_balonggandu_2023.xlsx'
df = read_excel(path_1)

df.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
# Decompose data untuk menemukan komponen tren dan musiman
result = seasonal_decompose(df['SMP'], model='additive', period=365)
result.plot()
plt.show()
```
