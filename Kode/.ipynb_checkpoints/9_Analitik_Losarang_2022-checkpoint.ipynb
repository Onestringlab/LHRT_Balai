### Kesimpulan
- **Outliers**: Semua distribusi menunjukkan adanya nilai outlier pada sisi kanan grafik, yang mungkin disebabkan oleh aktivitas spesifik atau hari-hari khusus.
- **Skewness**: Kebanyakan distribusi data condong ke kiri, terutama untuk kolom `SMP_MTR` dan `SMP_MBL`.
- **Pola Multimodal**: Distribusi `SMP_TRK` menunjukkan pola multimodal, yang bisa menunjukkan adanya dua atau lebih kelompok yang berbeda dalam penggunaan truk/bus.

```python
# Fungsi untuk menghapus outlier berdasarkan metode IQR
def remove_outliers(df, columns):
    # Membuat salinan data untuk menghindari perubahan langsung pada data asli
    df_clean = df.copy()
    
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Memfilter data agar hanya data dalam rentang yang diambil
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean
```

```python
# Contoh penggunaan untuk kolom 'SMP_MTR', 'SMP_MBL', 'SMP_TRK', dan 'SMP'
df_no_outliers = remove_outliers(df, ['SMP_MTR', 'SMP_MBL', 'SMP_TRK', 'SMP'])
```

```python
# Menentukan ukuran figure
plt.figure(figsize=(16, 12))

# Loop melalui tiap kolom numerik untuk membuat histogram dan KDE
for i, col in enumerate(['SMP_MTR', 'SMP_MBL', 'SMP_TRK', 'SMP']):
    plt.subplot(4, 2, i + 1)
    sns.histplot(df_no_outliers[col], kde=True)  # Menggunakan histogram dan KDE
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

Berikut adalah analisis hasil distribusi data setelah outlier dihapus:

1. **SMP_MTR (Sepeda Motor)**:
   - Distribusi `SMP_MTR` menjadi lebih simetris setelah outlier dihapus, dengan puncak frekuensi di sekitar 9,000.
   - Tidak lagi ada skewness yang kuat seperti sebelumnya, yang menunjukkan bahwa penghapusan outlier telah membantu mendekatkan distribusi ini ke bentuk yang lebih normal.
   - Rentang nilai lebih terbatas, berkisar antara 7,000 hingga 11,000.

2. **SMP_MBL (Mobil)**:
   - Distribusi `SMP_MBL` juga menjadi lebih simetris dan tidak lagi menunjukkan skewness ke kiri seperti sebelumnya.
   - Puncak distribusi terlihat di sekitar 6,000, dengan rentang yang lebih sempit, yaitu antara 5,000 hingga 7,500.
   - Dengan penghapusan outlier, distribusi ini menunjukkan pola yang lebih konsisten dan normal.

3. **SMP_TRK (Truk/Bus)**:
   - Distribusi `SMP_TRK` masih memiliki dua puncak (multimodal), meskipun bentuknya lebih seragam dan tidak ada nilai ekstrem yang menyebabkan deviasi besar.
   - Nilai terbanyak berada di sekitar 12,500 hingga 17,000, dengan rentang yang lebih terfokus antara 6,000 hingga 20,000.
   - Multimodalitas ini mungkin mengindikasikan adanya dua kelompok berbeda dalam data truk/bus, tetapi tanpa outlier, distribusinya menjadi lebih mudah dianalisis.

4. **SMP (Jumlah Semua Kendaraan)**:
   - Distribusi `SMP` menjadi lebih simetris dan menunjukkan puncak di sekitar 30,000.
   - Rentang data lebih sempit, yaitu antara 22,000 hingga 36,000, dan distribusi ini mendekati bentuk normal.
   - Penghapusan outlier telah membuat total jumlah kendaraan lebih terpusat, tanpa adanya hari-hari dengan jumlah kendaraan yang sangat ekstrem.

### Kesimpulan
- **Efek Penghapusan Outlier**: Penghapusan outlier membuat distribusi data pada masing-masing kolom menjadi lebih simetris dan mendekati bentuk normal, yang akan mempermudah analisis selanjutnya.
- **Multimodalitas pada `SMP_TRK`**: Distribusi `SMP_TRK` masih menunjukkan adanya dua kelompok, yang mungkin disebabkan oleh perbedaan penggunaan truk/bus pada hari-hari tertentu atau jenis kegiatan.
- **Lebih Mudah untuk Analisis Statistik**: Dengan distribusi yang lebih normal dan simetris, data ini akan lebih sesuai untuk teknik analisis statistik yang membutuhkan asumsi distribusi normal.

# 1. Segmentasi Berdasarkan Waktu (Hari Kerja vs Akhir Pekan)

```python
# Menambahkan kolom untuk menandai hari kerja dan akhir pekan
df_no_outliers['Segmen_Waktu'] = df_no_outliers['HARI'].apply(lambda x: 'Akhir Pekan' if x in ['Saturday', 'Sunday'] else 'Hari Kerja')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Segmen_Waktu', y='SMP_MTR', data=df_no_outliers)
plt.title('Segmentasi Jumlah SMP Sepeda Motor Berdasarkan Hari Kerja dan Akhir Pekan')
plt.xlabel('Segmen Waktu')
plt.ylabel('Jumlah SMP Sepeda Motor')
plt.show()
```

Dari grafik boxplot tersebut, berikut adalah analisis terkait segmentasi jumlah sepeda motor pada hari kerja dan akhir pekan:

1. **Median Lebih Tinggi pada Akhir Pekan**:
   - Median jumlah sepeda motor pada akhir pekan lebih tinggi dibandingkan hari kerja, menunjukkan bahwa rata-rata penggunaan sepeda motor cenderung meningkat saat akhir pekan. Hal ini mungkin disebabkan oleh aktivitas liburan atau kegiatan rekreasi yang lebih sering dilakukan pada hari Sabtu dan Minggu.

2. **Sebaran Data Lebih Luas pada Akhir Pekan**:
   - Rentang interkuartil (IQR) pada akhir pekan lebih lebar dibandingkan hari kerja. Hal ini menunjukkan variasi yang lebih besar dalam penggunaan sepeda motor saat akhir pekan, dengan beberapa hari yang memiliki jumlah penggunaan jauh lebih tinggi.
   - Hal ini mungkin mengindikasikan bahwa aktivitas pada akhir pekan lebih bervariasi, bergantung pada jenis kegiatan, musim, atau event khusus yang memengaruhi penggunaan sepeda motor.

3. **Outlier**:
   - Terdapat beberapa outlier pada kedua segmen (hari kerja dan akhir pekan), namun outlier lebih banyak muncul pada hari kerja. Ini bisa menunjukkan bahwa ada beberapa hari kerja dengan aktivitas yang cukup tinggi, mungkin karena adanya acara tertentu atau libur nasional yang jatuh pada hari kerja sehingga orang-orang lebih banyak menggunakan sepeda motor.
   - Untuk memahami outlier ini lebih jauh, analisis tambahan pada faktor-faktor eksternal seperti cuaca, libur nasional, atau event khusus di kota dapat dilakukan.

4. **Perbedaan Rata-rata Jumlah Penggunaan**:
   - Meskipun distribusi pada hari kerja cenderung lebih rendah dan stabil, ada peningkatan yang cukup signifikan pada akhir pekan. Ini menunjukkan bahwa aktivitas di akhir pekan berdampak langsung pada kenaikan jumlah penggunaan sepeda motor.

### Rekomendasi Tindakan
- **Segmentasi Lebih Lanjut Berdasarkan Musiman**: Untuk melihat pengaruh lebih jauh, Anda bisa melakukan segmentasi tambahan berdasarkan bulan atau musim tertentu (misalnya, libur panjang).
- **Analisis Khusus Hari Tertentu**: Melakukan analisis lebih dalam pada hari tertentu yang menunjukkan outlier dapat memberikan wawasan tambahan.
- **Prediksi Penggunaan**: Dengan pola yang lebih tinggi pada akhir pekan, model prediktif yang memperhitungkan hari dalam seminggu bisa lebih akurat dalam memprediksi jumlah penggunaan sepeda motor.

```python
# Menambahkan kolom untuk menandai hari kerja dan akhir pekan
df_no_outliers['Segmen_Waktu'] = df_no_outliers['HARI'].apply(lambda x: 'Akhir Pekan' if x in ['Saturday', 'Sunday'] else 'Hari Kerja')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Segmen_Waktu', y='SMP_MBL', data=df_no_outliers)
plt.title('Segmentasi Jumlah SMP Mobil Berdasarkan Hari Kerja dan Akhir Pekan')
plt.xlabel('Segmen Waktu')
plt.ylabel('Jumlah SMP Mobil')
plt.show()
```
