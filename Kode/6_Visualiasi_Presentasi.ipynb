# Losarang 2022

```python
# Data
data_terisi = 29893
data_kosong = 5147
total_data = 35040

# Save the pie chart as a PNG file
output_path = '../Visualisasi/losarang_2022/data_isi_kosong.png'

# Labels and sizes for the pie chart
labels = ['Data Terisi', 'Data Kosong']
sizes = [data_terisi, data_kosong]
colors = ['#F2E5BF', '#FD8B51']

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    sizes, labels=labels, colors=colors, autopct=lambda p: f'{p:.1f}%\n({int(p * total_data / 100):,})', 
    startangle=90, wedgeprops={'edgecolor': 'black'}, textprops={'color': 'black'})

plt.title('Data Losarang 2022')
plt.legend(wedges, [f'{label}: {value:,}' for label, value in zip(labels, sizes)], 
           title="Keterangan", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Save the chart to a file
plt.savefig(output_path, bbox_inches='tight', dpi=300)
```
